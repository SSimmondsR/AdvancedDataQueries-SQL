***Entrenenar mi modelo de machine learning utilizando train_df donde Churn es mi variable objetivo y los otros atributos son las caracteristicas.***

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
import pandas as pd

# train_df tiene la columna 'Churn', test_df no la tiene

# Paso 1: Entrenar un modelo de clasificación
# Suponiendo que 'features' es una lista de los nombres de las características en tus datos
features = ['AccountAge', 'MonthlyCharges', 'TotalCharges', 'PaperlessBilling',
       'MultiDeviceAccess', 'ViewingHoursPerWeek', 'AverageViewingDuration',
       'ContentDownloadsPerMonth', 'UserRating', 'SupportTicketsPerMonth',
       'WatchlistSize', 'ParentalControl', 'SubtitlesEnabled',
       'Gender__Female', 'Gender__Male', 'GenrePreference__Action',
       'GenrePreference__Comedy', 'GenrePreference__Drama',
       'GenrePreference__Fantasy', 'GenrePreference__Sci-Fi',
       'DeviceRegistered__Computer', 'DeviceRegistered__Mobile',
       'DeviceRegistered__TV', 'DeviceRegistered__Tablet', 'ContentType__Both',
       'ContentType__Movies', 'ContentType__TV Shows',
       'SubscriptionType__Basic', 'SubscriptionType__Premium',
       'SubscriptionType__Standard', 'PaymentMethod__Bank transfer',
       'PaymentMethod__Credit card', 'PaymentMethod__Electronic check',
       'PaymentMethod__Mailed check']
X_train = train_df_encoded[features]
y_train = train_df_encoded['Churn']

# Definicion y entrenamiento del modelo (en este caso, un RandomForestClassifier)
model = RandomForestClassifier()
model.fit(X_train, y_train)

# Paso 2: Predecir la variable objetivo en test_df
X_test = test_df_encoded[features]
predictions = model.predict(X_test)

# Paso 3: Agregar las predicciones como una columna en test_df
test_df_encoded['Churn'] = predictions

# Ahora test_df tiene una columna 'Churn' que representa las predicciones del modelo
```

```python
test_df_encoded
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score, confusion_matrix

# Etiquetas verdaderas del conjunto de datos de prueba
y_true = test_df_encoded['Churn']

# Precisión del modelo
accuracy = accuracy_score(y_true, predictions)
print("Accuracy:", accuracy)

# Precisión del modelo
precision = precision_score(y_true, predictions)
print("Precision:", precision)

# Recall del modelo
recall = recall_score(y_true, predictions)
print("Recall:", recall)

# F1-score del modelo
f1 = f1_score(y_true, predictions)
print("F1 Score:", f1)

# Matriz de confusión
conf_matrix = confusion_matrix(y_true, predictions)
print("Confusion Matrix:")
print(conf_matrix)

# ROC AUC Score
roc_auc = roc_auc_score(y_true, predictions)
print("ROC AUC Score:", roc_auc)
```
