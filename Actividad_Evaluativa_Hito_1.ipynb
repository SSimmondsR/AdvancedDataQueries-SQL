**Define qué es una red neuronal y describe sus componentes principales.**

Una red neuronal es un conjunto de algoritmos diseñados para reconocer patrones, estas redes interpretan datos sensoriales a traves de lentes de percepción similar a como los humanos y animales perciben el mundo.



Está compuesto por nodos, o neuronas artificales organizadas en capas (Capas de entrada, capas de salida, capas ocultas).




**Explica el proceso de backpropagation y su importancia en el entrenamiento de redes neuronales.**


El backpropagation es un método utilizado para calcular el gradiente de la función de pérdida respecto a cada peso en la red neuronal. Este proceso es fundamental para el aprendizaje de las redes, ya que permite ajustar los pesos de manera eficaz para minimizar el error de predicción. El procedimiento involucra dos fases principales:

Propagación hacia adelante - Los datos de entrada pasan a traves de la red para obtener obtener la salida, de izquierda a derecha

Retropropagación del error: Se cálcula el error entre el target y la capa de salida y se propaga de vuelta a traves de la red, actualizando los pesos según el gradiente de perdida.


**Describe tres funciones de activación comunes y sus características principales.**


ReLU: solo deja pasar valores positivos. Tiene un rango de salida de 	[0, ∞). Capas ocultas en la mayoría de modelos

Leaky ReLU: similar a ReLU, pero permite una pequeña pendiente para valores negativos. 	Rango de salida 	(-∞, ∞). Capas ocultas para evitar que neuronas "mueran"

tanh: da valores entre -1 y 1, útil cuando los datos están centrados. 	Capas ocultas; útil si los datos están centrados

Sigmoid: transforma cualquier número en un valor entre 0 y 1 (ideal para probabilidades). Capa de salida para clasificación binaria

ELU: mejora ReLU al tener una pendiente suave para valores negativos. Más estables o eficientes en ciertos casos. Rango de salida depende de la función

**1. Cargue de dataset**

```python
import kagglehub

# Download latest version
path = kagglehub.dataset_download("mlg-ulb/creditcardfraud")

print("Path to dataset files:", path)
```

```python
import pandas as pd

df = pd.read_csv('/content/creditcard.csv')
print(df.head())
df.info()
df = df.dropna(subset=['Class'])
```

**2. Preprocesamiento de datos**

```python
from re import X
## PIPELINE

from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer

from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler, OneHotEncoder

from sklearn.model_selection import train_test_split

## Creación de entrenamiento y testeo

X=df.drop('Class', axis=1)
y=df['Class']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=3035)
```

```python
## Numerical Transform

numerical_features=X.select_dtypes(include=['int64','float64']).columns

numerical_transformation=Pipeline(steps=[
    ('imputer',SimpleImputer(strategy='mean')),
    ('scaler',StandardScaler())
])

X_num_train=numerical_transformation.fit_transform(X_train[numerical_features])

pd.DataFrame(X_num_train)
```

```python
scaler=numerical_transformation.steps[1][1]
scaler.mean_
```

```python
## Integro con ColumnTransformer

preprocessor=ColumnTransformer(transformers=[
    ('Numerico',numerical_transformation,numerical_features),
    ])

X_train_prep=preprocessor.fit_transform(X_train)

X_train_prep=pd.DataFrame(X_train_prep)
X_train_prep.columns=numerical_features.tolist()
X_train_prep
```

```python
preprocessor
```

**3. Construya una red neuronal simple utilizando Keras.**

```python
# Construcción de red neuronal simple utilizando Keras

# Importaciones de librerias básicas
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
from matplotlib.colors import ListedColormap


# TensorFlow y Keras
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers import Adam

# Configuración
np.random.seed(42)
tf.random.set_seed(42)

print(f"TensorFlow version: {tf.__version__}")
print(f"Eager execution: {tf.executing_eagerly()}")
```

```python
simple_model=Sequential()
simple_model.add(Dense(64,activation='tanh'))
simple_model.add(Dense(32,activation='tanh'))
simple_model.add(Dense(1,activation='sigmoid'))

simple_model.compile(loss='binary_crossentropy',optimizer=Adam(learning_rate=0.001),metrics=['accuracy'])

simple_model.fit(X_train_prep,y_train,epochs=50,verbose=1)
```

Definiendo 50 epocas para el modelo se puede deducir que en la epoca 17, ya no hay mejoras significativas.

El modelo ya aprendió lo que debía aprender y seguir entrenando solo consume tiempo/computo y puede causar overfitting.

```python
## Evaluar el modelo
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

y_pred=simple_model.predict(X_train_prep)

y_pred=(y_pred>0.5)

print(accuracy_score(y_train,y_pred))
print(confusion_matrix(y_train,y_pred))
print(classification_report(y_train,y_pred))
```

```python
X_test_prep=preprocessor.transform(X_test)

y_pred=simple_model.predict(X_test_prep)

y_pred=(y_pred>0.5)

print(classification_report(y_test,y_pred))
```
